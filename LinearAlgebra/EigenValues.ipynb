### Eigenvectors and Eigenvalues

Given a square matrix $\mathbf{A}\in\mathbb{R}^{n,n}$, we say that $\lambda\in\mathbb{C}$ is an eigenvalue of $\mathbf{A}$ and vector $\mathbf{v}\in\mathbb{R}^{n}$ is the corresponding eigen vector if $$\mathbf{Av} = \lambda \mathbf{v},  v!=0 $$
Intuitively, this definition means that multiplying A by the vector v results in a new vector
that points in the same direction as v, but scaled by a factor $\lambda$. If the eigenvalue is greater than one, the corresponding Avᵢ will expand. If it is smaller than one, it will shrink.
<br/>
<br/>
Also note that for any eigenvector x ∈ $\mathbb{C}^{n}$, and scalar c ∈ $\mathbb{C}$,  A(cx) = cAx = cλx = λ(cx), so cx is also an eigenvector. For this reason when we talk about “the” eigenvector associated with λ, we usually assume that the eigenvector is normalized to have length 1
<br/>
<br/>
We can rewrite the equation above to state that (λ, v) is an eigenvalue-eigenvector pair
of A if,    <br/>(λI − A)v = 0, v != 0. <br/><br/>
But (λI − A)v = 0 has a non-zero solution to v if and only if (λI − A) has a non-empty
nullspace, which is only the case if (λI − A) is singular (non-invertible), i.e., <br/><br/>
|(λI − A)| = 0 or det((λI − A)) = 0
<br/><br/>
Intuition behind the condition:
- An invertible matrix (det(A) ≠ 0) indicates a unique solution for the system of equations. This unique solution will always be the zero vector (x = 0) if the constant vector (b) is also zero.
- When A is singular (det(A) = 0), it suggests that the rows or columns of the matrix are dependent. This dependence can lead to scenarios where some elements in the unknowns vector (x) can be non-zero while still satisfying the equation Ax=0.

#### Finding Eigenvalues and Eigenvectors

To find the eigenvalues and eigenvectors of a matrix $\mathbf{A}$, follow these steps:

1. **Characteristic Equation**:
   - Form the characteristic equation by subtracting $\lambda \mathbf{I}$ from $\mathbf{A}$ and setting the determinant to zero:
     $$
     \text{det}(\mathbf{A} - \lambda \mathbf{I}) = 0
     $$
   - Here, $\mathbf{I}$ is the identity matrix of the same size as $\mathbf{A}$.

2. **Solve for Eigenvalues**:
   - Solve the characteristic equation for $\lambda$ to find the eigenvalues.

3. **Find Eigenvectors**:
   - For each eigenvalue $\lambda$, solve the equation $(\mathbf{A} - \lambda \mathbf{I}) \mathbf{v} = \mathbf{0}$ to find the corresponding eigenvector(s) $\mathbf{v}$.

#### Conditions for Existence of Eigenvectors

1. **Square Matrix**:
   - Only square matrices (i.e., matrices with the same number of rows and columns) can have eigenvectors and eigenvalues. Thus, $\mathbf{A}$ must be an $n \times n$ matrix.

2. **Characteristic Polynomial**:
   - The eigenvalues of $\mathbf{A}$ are the roots of the characteristic polynomial, which is obtained from the characteristic equation:
     $$
     \text{det}(\mathbf{A} - \lambda \mathbf{I}) = 0
     $$
     where $\mathbf{I}$ is the $n \times n$ identity matrix. The matrix $\mathbf{A} - \lambda \mathbf{I}$ is called the characteristic matrix.

3. **Non-Trivial Solution**:
   - For $\mathbf{v}$ to be an eigenvector, the equation $(\mathbf{A} - \lambda \mathbf{I}) \mathbf{v} = \mathbf{0}$ must have a non-trivial solution (i.e., $\mathbf{v} \neq \mathbf{0}$). This requires that $\mathbf{A} - \lambda \mathbf{I}$ is singular (non-invertible), meaning its determinant is zero:
     $$
     \text{det}(\mathbf{A} - \lambda \mathbf{I}) = 0
     $$

4. **Existence of Eigenvalues**:
   - If the characteristic polynomial has at least one real or complex root, then $\mathbf{A}$ has at least one eigenvalue and corresponding eigenvector. This is guaranteed for any square matrix because every polynomial of degree $n$ has exactly $n$ roots in the complex number system (including multiplicities), according to the Fundamental Theorem of Algebra.

5. **Diagonalizability**:
   - A matrix $\mathbf{A}$ is diagonalizable if there are $n$ linearly independent eigenvectors. This means $\mathbf{A}$ can be expressed as:
     $$
     \mathbf{A} = \mathbf{P} \mathbf{D} \mathbf{P}^{-1}
     $$
     where $\mathbf{D}$ is a diagonal matrix containing the eigenvalues of $\mathbf{A}$, and $\mathbf{P}$ is a matrix whose columns are the corresponding eigenvectors. However, even non-diagonalizable matrices have eigenvectors, though they may not span the entire space.

#### Properties Related to Eigenvectors

1. **Distinct Eigenvalues**:
   - If a matrix $\mathbf{A}$ has $n$ distinct eigenvalues, then it has $n$ linearly independent eigenvectors, and hence it is diagonalizable. Eigenvectors corresponding to distinct eigenvalues are linearly independent.

2. **Geometric Multiplicity**:
   - The geometric multiplicity of an eigenvalue $\lambda$ is the dimension of its eigenspace (the set of all eigenvectors corresponding to $\lambda$, along with the zero vector). This is always less than or equal to the algebraic multiplicity (the number of times $\lambda$ appears as a root of the characteristic polynomial).

3. **Symmetric Matrices**:
   - A symmetric matrix (i.e., $\mathbf{A} = \mathbf{A}^\top$) always has real eigenvalues and is always diagonalizable. The eigenvectors of a symmetric matrix corresponding to distinct eigenvalues are orthogonal.

4. **Determinant and Trace**:
   - The determinant of a square matrix $\mathbf{A}$ is the product of its eigenvalues:
     $$
     \text{det}(\mathbf{A}) = \prod_{i=1}^n \lambda_i
     $$
   - The trace of a square matrix $\mathbf{A}$ is the sum of its eigenvalues:
     $$
     \text{tr}(\mathbf{A}) = \sum_{i=1}^n \lambda_i
     $$

5. **Eigenvalues of Inverse**:
   - If $\mathbf{A}$ is invertible, the eigenvalues of $\mathbf{A}^{-1}$ are the reciprocals of the eigenvalues of $\mathbf{A}$:
     $$
     \text{If } \lambda \text{ is an eigenvalue of } \mathbf{A}, \text{ then } \frac{1}{\lambda} \text{ is an eigenvalue of } \mathbf{A}^{-1}
     $$

### Eigen Decomposition

Eigen decomposition is the process of decomposing a square matrix $\mathbf{A}$ into a product of three matrices:
$$
\mathbf{A} = \mathbf{P} \mathbf{D} \mathbf{P}^{-1}
$$
where:
- $\mathbf{P}$ is a matrix whose columns are the eigenvectors of $\mathbf{A}$,
- $\mathbf{D}$ is a diagonal matrix whose diagonal elements are the eigenvalues of $\mathbf{A}$,
- $\mathbf{P}^{-1}$ is the inverse of the matrix $\mathbf{P}$.

<u> Conditions for Eigen Decomposition </u>

For a matrix $\mathbf{A}$ to be diagonalizable (i.e., it can be decomposed in this way), it must have $n$ linearly independent eigenvectors, where $n$ is the size of the matrix. This is equivalent to saying that the algebraic multiplicity of each eigenvalue equals its geometric multiplicity.

<u> Steps for Eigen Decomposition </u>

1. **Find the Eigenvalues**:
   - Solve the characteristic equation $\text{det}(\mathbf{A} - \lambda \mathbf{I}) = 0$ to find the eigenvalues of $\mathbf{A}$.

2. **Find the Eigenvectors**:
   - For each eigenvalue $\lambda$, solve the equation $(\mathbf{A} - \lambda \mathbf{I}) \mathbf{v} = \mathbf{0}$ to find the corresponding eigenvectors.

3. **Form the Matrices $\mathbf{P}$ and $\mathbf{D}$**:
   - Construct $\mathbf{P}$ using the eigenvectors as columns.
   - Construct $\mathbf{D}$ as a diagonal matrix with the corresponding eigenvalues on the diagonal.

#### Example

Consider the matrix:
$$
\mathbf{A} = \begin{pmatrix}
5 & 4 \\
1 & 2
\end{pmatrix}
$$

### Step 1: Find the Eigenvalues

Form the characteristic equation:
$$
\text{det}(\mathbf{A} - \lambda \mathbf{I}) = \text{det}\begin{pmatrix}
5 - \lambda & 4 \\
1 & 2 - \lambda
\end{pmatrix} = 0
$$

Calculate the determinant:
$$
(5 - \lambda)(2 - \lambda) - 4 \cdot 1 = \lambda^2 - 7\lambda + 6 = 0
$$

Solve the quadratic equation for $\lambda$:
$$
\lambda^2 - 7\lambda + 6 = 0
$$

The roots are:
$$
\lambda_1 = 6 \quad \text{and} \quad \lambda_2 = 1
$$

### Step 2: Find the Eigenvectors

#### For $\lambda = 6$:

Solve $(\mathbf{A} - 6 \mathbf{I}) \mathbf{v} = \mathbf{0}$:
$$
\begin{pmatrix}
-1 & 4 \\
1 & -4
\end{pmatrix} \begin{pmatrix}
v_1 \\
v_2
\end{pmatrix} = \begin{pmatrix}
0 \\
0
\end{pmatrix}
$$

This system simplifies to:
$$
- v_1 + 4 v_2 = 0 \Rightarrow v_1 = 4 v_2
$$

The eigenvector corresponding to $\lambda = 6$ is:
$$
\mathbf{v}_1 = \begin{pmatrix}
4 \\
1
\end{pmatrix}
$$

#### For $\lambda = 1$:

Solve $(\mathbf{A} - 1 \mathbf{I}) \mathbf{v} = \mathbf{0}$:
$$
\begin{pmatrix}
4 & 4 \\
1 & 1
\end{pmatrix} \begin{pmatrix}
v_1 \\
v_2
\end{pmatrix} = \begin{pmatrix}
0 \\
0
\end{pmatrix}
$$

This system simplifies to:
$$
4 v_1 + 4 v_2 = 0 \Rightarrow v_1 = - v_2
$$

The eigenvector corresponding to $\lambda = 1$ is:
$$
\mathbf{v}_2 = \begin{pmatrix}
1 \\
-1
\end{pmatrix}
$$

### Step 3: Form the Matrices $\mathbf{P}$ and $\mathbf{D}$

Construct $\mathbf{P}$ using the eigenvectors as columns:
$$
\mathbf{P} = \begin{pmatrix}
4 & 1 \\
1 & -1
\end{pmatrix}
$$

Construct $\mathbf{D}$ as a diagonal matrix with the eigenvalues on the diagonal:
$$
\mathbf{D} = \begin{pmatrix}
6 & 0 \\
0 & 1
\end{pmatrix}
$$

### Verify the Decomposition

To verify the eigen decomposition, check that:
$$
\mathbf{A} = \mathbf{P} \mathbf{D} \mathbf{P}^{-1}
$$

First, find $\mathbf{P}^{-1}$:
$$
\mathbf{P}^{-1} = \frac{1}{4(-1) - 1(1)} \begin{pmatrix}
-1 & -1 \\
-1 & 4
\end{pmatrix} = \frac{1}{-5} \begin{pmatrix}
-1 & -1 \\
-1 & 4
\end{pmatrix} = \begin{pmatrix}
\frac{1}{5} & \frac{1}{5} \\
\frac{1}{5} & -\frac{4}{5}
\end{pmatrix}
$$

Now, multiply $\mathbf{P}$, $\mathbf{D}$, and $\mathbf{P}^{-1}$:
$$
\mathbf{P} \mathbf{D} = \begin{pmatrix}
4 & 1 \\
1 & -1
\end{pmatrix} \begin{pmatrix}
6 & 0 \\
0 & 1
\end{pmatrix} = \begin{pmatrix}
24 & 1 \\
6 & -1
\end{pmatrix}
$$

$$
\mathbf{P} \mathbf{D} \mathbf{P}^{-1} = \begin{pmatrix}
24 & 1 \\
6 & -1
\end{pmatrix} \begin{pmatrix}
\frac{1}{5} & \frac{1}{5} \\
\frac{1}{5} & -\frac{4}{5}
\end{pmatrix} = \begin{pmatrix}
24(\frac{1}{5}) + 1(\frac{1}{5}) & 24(\frac{1}{5}) + 1(-\frac{4}{5}) \\
6(\frac{1}{5}) - 1(\frac{1}{5}) & 6(\frac{1}{5}) - 1(-\frac{4}{5})
\end{pmatrix}
= \begin{pmatrix}
5 & 4 \\
1 & 2
\end{pmatrix}
= \mathbf{A}
$$

Hence, the eigen decomposition is verified.
